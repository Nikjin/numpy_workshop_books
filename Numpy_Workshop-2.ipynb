# **1.Mathematical Functions**

## **1.1 Rounding Decimals**

**Truncation**

Remove the decimals, and return the float number closest to zero. 

```
arr = np.trunc([-3.1666, 3.6667])

print(arr)
```

**Rounding**

The around() function increments preceding digit or decimal by 1 if >=5 else do nothing.

```
arr = np.around(3.16666, 2)

print(arr)
```

**Floor**

The floor() function rounds off decimal to nearest lower integer.

image.png

```
arr = np.floor([-3.1666, 3.6667])

print(arr)
```

**Ceil**

The ceil() function rounds off decimal to nearest upper integer.

```
arr = np.ceil([-3.1666, 3.6667])

print(arr)
```

## **1.2 Logs**

NumPy provides functions to perform log at the base 2, e and 10.

**Log at Base 2**

Use the log2() function to perform log at the base 2.

```
arr = np.arange(1, 10)

print(np.log2(arr))
```

**Log at Base 10**

Use the log10() function to perform log at the base 10.

```
arr = np.arange(1, 10)

print(np.log10(arr))
```

**Natural Log, or Log at Base e**

Use the log() function to perform log at the base e.

```
arr = np.arange(1, 10)

print(np.log(arr))
```

# **2.Trigonometric Functions**

NumPy has standard trigonometric functions which return trigonometric ratios for a given angle in radians.

## **2.1 Trigonometric ratios**

```
x = np.sin(np.pi/2)

print(x)
```

```
print("Cosine 90: ")
print(np.cos(np.pi/2))

print("\nTan 90: ")
print(np.tan(np.pi/2))
```

```
a = np.array([np.pi/2, np.pi/3, np.pi/4, np.pi/5])

print("Sine ratios:\n", np.sin(a))

print("\nCosine ratios:\n", np.cos(a))

print("\nTan ratios:\n", np.tan(a))
```

## **2.2 Degrees and Radians Conversions**

By default all of the trigonometric functions take radians as parameters but we can convert radians to degrees and vice versa

**Degrees to Radians**

```
arr = np.array([90, 180, 270, 360])

x = np.deg2rad(arr)
# we can also use np.radians()

print(x)
```

**Radians To Degrees**

```
arr = np.array([np.pi/2, np.pi, 1.5*np.pi, 2*np.pi])

x = np.rad2deg(arr)
# We can also use np.degrees()

print(x)
```

## **2.3 Finding Angles**

NumPy provides arcsin(), arccos() and arctan() that produce radian values for corresponding sin, cos and tan values given

```
print("Sine Inverse 1.0\n")
x = np.arcsin(1.0)

print("In Radians",x)

print("\nIn Degrees",np.rad2deg(x))
```

```
print("Cosine Inverse 1.0")

x = np.arccos(1.0)

print("\nIn Radians",x)

print("In Degrees",np.rad2deg(x))


print("\n\nTan Inverse 1.0")

x = np.arctan(1.0)

print("\nIn Radians",x)

print("In Degrees",np.rad2deg(x))
```

```
arr = np.array([1, -1, 0.1])

x = np.arcsin(arr)

print(np.rad2deg(x))
```

## **2.4 Finding Hypotenues**

NumPy provides the hypot() function that takes the base and perpendicular values and produces hypotenues based on pythagoras theorem.

```
base = 3
perp = 4

x = np.hypot(base, perp)

print(x)
```

# **3.Statistical Functions**

NumPy has quite a few useful statistical functions for finding minimum, maximum, percentile standard deviation and variance, etc. from the given elements in the array.

**Example**

Here are the marks for 50 students.

82, 57, 85,  6, 49, 77, 32, 85, 52, 90, 83, 31, 87, 11, 62, 31, 80, 45, 16, 66, 27, 27, 50, 32, 87, 26, 51, 29, 81, 83, 58, 45, 52, 51, 93, 86, 77, 39, 56, 97, 40, 45, 96, 45, 58, 45, 54,  6,  5, 48.

Analyze the data and gain **statistical insights** from it.

```
result = np.array([82, 57, 85,  6, 49, 77, 32, 85, 52, 90, 83, 31, 87, 11, 62, 31, 80, 45, 16, 66, 
                   27, 27, 50, 32, 87, 26, 51, 29, 81, 83, 58, 45, 52, 51, 93, 86, 77, 39, 56, 97, 
                   40, 45, 96, 45, 58, 45, 54,  6,  5, 48])
```

Find what is the **minimum** marks

```
np.amin(result)
```

Find what is the **maximum** marks

```
np.amax(result)
```

Find what is the marks obtained on an **average**(mean)

```
np.mean(result)
```

What is the **central value**(median) of the result.

```
np.median(result)
```

Find the mark of student who secured 80th **Percentile Rank**.

```
np.percentile(result,80)
```

How much **dispersed**(Standard Deviation) the marks are in the result.

```
np.std(result)
```

How much random marks are dispersed from their expected marks

(Varience)

```
np.var(result)
```

# **5.Linear Algebra**

The Linear Algebra module of NumPy offers various methods to apply linear algebra on any numpy array.

```
A = np.array([[1, 1, 1],
              [1, 0, -1],
              [0, 1, -3]])
```

Rank of the Matrix

```
np.linalg.matrix_rank(A)
```

Determinant of a Matrix

```
np.linalg.det(A)
```

Inverse of a Matrix

```
np.linalg.inv(A)
```

Power to a Matrix

```
np.linalg.matrix_power(A, 3)
```

Eigen Values and Eigen vectors

```
eval,evecs = np.linalg.eig(A)

print("Eigen Values:\n",eval)

print("\nEigen Vectors:\n",evecs)
```

**Solve a system of Linear Equations**

If I bought

Day 1: 0 oranges, 3 bananas, and 4 apples and it cost ₹32.

Day 2: 3 oranges, 2 bananas, and 1 apple and it cost ₹19.

Day 3: 3 oranges, 3 bananas and 3 apples and it cost ₹33. 

How much each fruit cost?

```
A = np.array([[0,3,4],
              [3,2,1],
              [3,3,3]])

B = np.array([32,19,33])
```

```
np.linalg.solve(A,B)
```

# **6.Random Module**

NumPy offers the random module to work with random numbers.


```
from numpy import random
```

Generate Random Integers

```
random.randint(5,high=7,size=(2,4))
```

Generate Random Floats

```
#Range between 0 and 1.
random.rand(5,3)
```

Select Random numbers from the given values

```
random.choice([3, 5, 7, 9, 11, 15, 18],size=(2,3))
```

Select Random numbers based on probabilities from the given values

```
random.choice([9, 11, 15, 18], p=[0.2, 0.4, 0.3, 0.1], size=(3,4))
```

**Shuffling Arrays**

Shuffle means changing arrangement of elements in-place. i.e. in the array itself.

```
A = np.array([1, 2, 3, 4, 5])

random.shuffle(A)

print(A)
```

Similar is permuatation function, the difference is that shuffle makes changes to original array, while permuation returns a re-arranged array

```
random.permutation([1,2,3,4,5,6])
```

# **8.Speed Comparison**

## **Summation**

```
X = np.random.randint(100,200,(1000))

print(X.shape)
```

```
def mySum(X):
  result = 0
  for element in X:
    result += element
```

```
%timeit mySum(X)
```

```
%timeit sum(X)
```

```
%timeit np.sum(X)
```

## **Power Of Matrix**

```
A = np.random.randint(100,200,(100,100))
```

```
def myMatrixPower(A,N):
  X = A
  for i in range(N-1):
    X = X @ A
```

```
%timeit myMatrixPower(A,500)
```

```
%timeit np.linalg.matrix_power(A, 3)
```

## **Matrix Multiplication**

```
A = np.random.randint(20,80,(100,100))
B = np.random.randint(20,80,(100,100))

print(A.shape,B.shape,sep="\n\n")
```

```
def myMatrixMultiplier(A,B):

  result = [[sum(a * b for a, b in zip(A_row, B_col))  
                        for B_col in zip(*B)] 
                                for A_row in A]
```

```
%timeit np.matmul(A,B)
```

```
%timeit myMatrixMultiplier(A,B)
```
